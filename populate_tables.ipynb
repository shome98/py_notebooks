```python
import asyncio
import requests
import json
```

```python
with open("random_users25jun25.json") as f:
    users=json.load(f)

url="http://127.0.0.1:8000/api/register"
for user in users:
    response=requests.post(url,json=user)
    print(f"Registered {user['email']}->{response.json()}")
```

```python
with open("random_users25jun25.json") as f:
    users=json.load(f)

url="http://127.0.0.1:8000/api/login"
for user in users:
    response=requests.post(url,json={'email':user['email'],'password':user['password']})
    print(f"Registered {user['email']}->{response.json()}")
```

```python
with open("random_users25jun25.json") as f:
    users = json.load(f)

url = "http://127.0.0.1:8000/api/login"

for user in users:
    print(f"Trying {user['email']}...")
    response = requests.post(url, json={'email': user['email'], 'password': user['password']})
    
    # Check if response is valid JSON before decoding
    try:
        data = response.json()
        print(f"Success: {data}")
    except requests.exceptions.JSONDecodeError:
        print(f"Failed for {user['email']} - Invalid JSON response")
        print("Raw response:", response.text[:500])
```
