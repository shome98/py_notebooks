```python
import requests
import json
import os

# --- Configuration ---
API_ENDPOINT = "http://localhost:9876/api/v1/products/products/"
DATA_FILE_PATH = os.path.join(os.path.expanduser("~"), "Desktop", "products_data.json")
IMAGE_FOLDER_PATH = os.path.join(os.path.expanduser("~"), "Desktop", "product_images")
FREE_KEY = "67e18653bac618efa9031920"  # Replace with your actual free key

HEADERS = {
    "free_key": FREE_KEY
}

def add_product_with_image(product_data, image_path):
    """Posts a single product with its associated image."""
    data = {
        "name": product_data.get("name"),
        "category": product_data.get("category"),
        "sku": product_data.get("sku"),
        "price": product_data.get("price"),
        "quantityInStock": product_data.get("quantityInStock", 0),
        "userId": product_data.get("userId"),
        "description": product_data.get("description")  # Include description
        # Add other non-image fields
    }
    files = [('images', (os.path.basename(image_path), open(image_path, 'rb')))]
    try:
        response = requests.post(API_ENDPOINT, headers=HEADERS, data=data, files=files)
        response.raise_for_status()
        print(f"Product '{data.get('name', 'Unnamed')}' added. Response: {response.json()}")
    except requests.exceptions.RequestException as e:
        print(f"Error adding product '{data.get('name', 'Unnamed')}': {e}")
        if hasattr(response, 'status_code'):
            print(f"Status Code: {response.status_code}")
        if hasattr(response, 'text'):
            print(f"Body: {response.text}")
    finally:
        for name, file_tuple in files:
            if isinstance(file_tuple, tuple) and hasattr(file_tuple[1], 'close'):
                file_tuple[1].close()

def main():
    try:
        with open(DATA_FILE_PATH, 'r') as f:
            products = json.load(f)
    except FileNotFoundError:
        print(f"Error: {DATA_FILE_PATH} not found.")
        return
    except json.JSONDecodeError:
        print(f"Error: Invalid JSON in {DATA_FILE_PATH}.")
        return

    image_files = sorted([os.path.join(IMAGE_FOLDER_PATH, f) for f in os.listdir(IMAGE_FOLDER_PATH) if os.path.isfile(os.path.join(IMAGE_FOLDER_PATH, f))])

    num_products = len(products)
    num_images = len(image_files)
    count = min(num_products, num_images)

    print(f"Attempting to add {count} products with images...")

    for i in range(count):
        product = products[i]
        image_path = image_files[i]
        add_product_with_image(product, image_path)

    print("Finished processing products and images.")

if __name__ == "__main__":
    main()
```
